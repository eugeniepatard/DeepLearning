```python
import pandas as pd

# Load attribute labels
attributes = pd.read_csv('/content/celeba-dataset/list_attr_celeba.csv')
attributes = attributes['Male']
attributes.head(5)
```

```python
# Load partition data
partitions = pd.read_csv('/content/celeba-dataset/list_eval_partition.csv')
partitions
```

```python
data = pd.concat([partitions, attributes], axis=1)
data.head(5)
```

```python
class CelebaDataset(Dataset):
  def __init__(self, data, img_dir, transform=None):
    super().__init__()
    self.data = data
    self.img_dir = img_dir
    self.transform = transform

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx):
    row = self.data.iloc[idx]
    image_name = row["image_id"]
    path = os.path.join(self.img_dir, image_name)
    image = Image.open(path).convert("RGB")
    label = row["Male"]

    if self.transform:
            image = self.transform(image)

    return image, label
```

```python
train_data = data[data["partition"] == 0][:3000]
val_data = data[data["partition"] == 1][:300]
test_data = data[data["partition"] == 2][:300]

img_dir = "/content/celeba-dataset/img_align_celeba/img_align_celeba"
```

```python
train_data.shape
```

```python
transform = transforms.Compose([transforms.Resize((128, 128)),
                                transforms.ToTensor()])

train_dataset = CelebaDataset(train_data, img_dir, transform)
val_dataset = CelebaDataset(val_data, img_dir, transform=transform)
test_dataset = CelebaDataset(test_data, img_dir, transform=transform)
```

```python
batch_size = 64

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Initialize mean and std
mean = torch.zeros(3)
std = torch.zeros(3)
n_samples = 0

# Calculate mean and std
for images, _ in tqdm(train_loader):
    batch_samples = images.size(0)  # Batch size
    n_samples += batch_samples
    mean += images.mean([0, 2, 3]) * batch_samples
    std += images.std([0, 2, 3]) * batch_samples

mean /= n_samples
std /= n_samples

print(f"Mean: {mean}")
print(f"Std: {std}")
```

```python
import torch.nn as nn

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2)
        )

        self.flatten = nn.Flatten()
        self.fc = nn.Sequential(
            nn.Linear(64 * 32 * 32, 128),  # Will be updated dynamically
            nn.ReLU(),
            nn.Linear(128, 1),
            nn.Sigmoid()  # For binary classification
        )

    def forward(self, x):
        x = self.conv(x)
        x = self.flatten(x)
        x = self.fc(x)
        return x
```

```python
from sklearn.metrics import f1_score
import torch.nn.functional as F
import numpy as np

def train(model, optimiser, criterion, loader):
    model.train()
    train_accuracy, train_loss = 0, 0
    for x, labels in loader:
        # Move data to the same device as the model
        x, labels = x.to(device), labels.to(device)

        labels = labels.float()

        optimiser.zero_grad()
        output = model(x)

        # Compute loss
        loss = criterion(output.squeeze(), labels)
        loss.backward()
        optimiser.step()

        # Accumulate loss
        train_loss += loss.item() * x.size(0)

        # Compute predictions and F1 score
        y_pred = (torch.sigmoid(output.squeeze()) > 0.5).long()
        train_accuracy += f1_score(labels.cpu().numpy(), y_pred.cpu().numpy(), average='weighted') * x.size(0)

    return train_accuracy / len(loader.dataset), train_loss / len(loader.dataset)

def validate(model, criterion, data_loader):
    model.eval()
    validation_loss, validation_accuracy = 0., 0.
    for x, labels in data_loader:
        with torch.no_grad():
            x, labels = x.to(device), labels.to(device)
            labels = labels.float()
            output = model(x)

            # Compute loss
            loss = criterion(output.squeeze(), labels)
            validation_loss += loss.item() * x.size(0)

            # Compute predictions and F1 score
            y_pred = (torch.sigmoid(output.squeeze()) > 0.5).long()
            validation_accuracy += f1_score(labels.cpu().numpy(), y_pred.cpu().numpy(), average='weighted') * x.size(0)

    return validation_loss / len(data_loader.dataset), validation_accuracy / len(data_loader.dataset)


def evaluate(model, data_loader):
    model.eval()
    ys, y_preds = [], []
    for X, y in data_loader:
        with torch.no_grad():
            a2 = model(X.view(-1, 3, 224, 224))
            y_pred = F.log_softmax(a2, dim=1).max(1)[1]
            ys.append(y.cpu().numpy())
            y_preds.append(y_pred.cpu().numpy())

    return np.concatenate(y_preds, 0),  np.concatenate(ys, 0)
```
